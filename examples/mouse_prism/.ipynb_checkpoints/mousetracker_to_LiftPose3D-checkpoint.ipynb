```python
files = sorted(glob.glob(data_dir + '*.mat'))
videos = sorted(glob.glob(data_dir + '*.avi'))

for idx in range(len(files)):
    print(files[idx])
     
    data = loadmat(files[idx])
    all_points = np.concatenate((data['final_tracks'],data['tracks_tail']),axis=1)
    
    #check first frame where mouse is in field of view
    n = all_points.shape[2]
    for i in range(n):
        if (~np.isnan(all_points[:,:6,i])).all():
            break
      
    index = np.arange(i,n)
                            
    #add point between tail root and nose tip
    all_points[:,6,:] = (all_points[:,4,:] + all_points[:,5,:])/2
    all_points[2,6,:] = data['mirror_line'][0][0]
    
    #discard tail points, keep only body points
    all_points = all_points[:,:7,index]
    
    #change to DF3D format    
    bottom = np.stack((all_points[0,:,:],all_points[1,:,:]),axis=2)
    side = np.stack((all_points[0,:,:],all_points[2,:,:]),axis=2)
    points2d = np.stack((bottom,side),axis=0)
    points3d = np.swapaxes(all_points,0,2)
    points2d = np.swapaxes(points2d,1,2)
    
    #set NaNs to zero as otherwise training will fail
    points2d[np.isnan(points2d)] = 0
    points3d[np.isnan(points3d)] = 0
    
    #align mice pointing left
    if files[idx][-6] == 'L':
        cos, sin = np.cos(np.pi), np.sin(np.pi)
        R = np.array(((cos, -sin), (sin, cos)))
        for i in range(points2d.shape[2]):
            tmp = points2d[0,i,:,:].copy()
            c = tmp[6,:]
            tmp = np.matmul(tmp-c,R) + c
            points2d[0,i,:,:] = tmp
        
    #good keypoints
    good_keypts = ~np.isnan(points3d).any(2)
        
    poses = {'points3d': points3d, 
             'points2d':points2d, 
             'good_keypts': good_keypts,
             'index': index,
             'mirror_line':data['mirror_line']
             'distort': data['IDX']}
        
    pickle.dump(poses,open(files[idx][:-7] + '.pkl', 'wb'))
```
